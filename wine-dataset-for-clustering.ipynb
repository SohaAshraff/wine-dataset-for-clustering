```python
def calc_mean(data):
    mean = np.mean(data, axis=0)  
    return mean
Mean = calc_mean(data)
Mean 
```

```python
def calc_covariance_matrix(data, mean):
   
    shifted_data = data - mean  
    print("Shifted Feature Vectors (F - m):\n", shifted_data)
    
    covariance_matrix = np.cov(shifted_data.T)  
    return covariance_matrix, shifted_data
cov_matrix, shifted_data = calc_covariance_matrix(data, Mean)
```

```python
print("Covariance Matrix:\n", cov_matrix)
```

```python
def calculate_eign_value(cov_matrix):
    
    coeffs = np.poly(cov_matrix)
    
    eign_values = np.roots(coeffs)
    return eign_values

eign_values = calculate_eign_value(cov_matrix)
print(eign_values.shape)
print(eign_values)
```

```python
def calculate_eign_vectors(cov_matrix, eign_values):
    n = cov_matrix.shape[0]
    eig_vectors = []
    for eig_val in eign_values:
        eig_vec_matrix = eig_val * np.eye(n) - cov_matrix
        # use SVD (singular value decomposition) to solve the equation of eig_vec_matrix to get eign vector
        _, _, vh = np.linalg.svd(eig_vec_matrix)
        eig_vector = vh[-1] # last row in vh is the eign vector we seek
        eig_vector = eig_vector / np.linalg.norm(eig_vector)
        eig_vectors.append(eig_vector)
    eign_vectors = np.array(eig_vectors).T
    return eign_vectors
eign_vectors = calculate_eign_vectors(cov_matrix, eign_values)

eign_vectors
```

```python
def calculate_explained_variance(eigenvalues):
    total_variance = np.sum(eigenvalues)
    explained_variance = eigenvalues / total_variance
    return explained_variance

explained_variance = calculate_explained_variance(eign_values)
print("Explained Variance:", explained_variance)
```

```python
def construct_q_matrix(eigenvalues, eigenvectors, n_components):
    
    idx = np.argsort(eigenvalues)[::-1]
    
 
    eigenvectors_sorted = eigenvectors[:, idx]
    
  
    eigenvectors_normalized = eigenvectors_sorted / np.linalg.norm(eigenvectors_sorted, axis=0)
    
    
    q_matrix = eigenvectors_normalized[:, :n_components]
    return q_matrix

n_components = 5# Example: reduce to 2 components
q_matrix = construct_q_matrix(eign_values, eign_vectors, n_components)
print("Q Matrix:\n", q_matrix)
```

```python
def transform(q, shifted_data):
    transformed_data = np.dot(shifted_data, q)
    return transformed_data

transformed_data = transform(q_matrix, shifted_data)
print("Transformed Data :\n", transformed_data)
```

```python
transformed_data.shape
```

```python
def inverse_transform(q, transformed_data, mean):
    mean_reshaped = np.tile(mean, (transformed_data.shape[0], 1))  
    original_data = np.dot( transformed_data ,q.T) + mean_reshaped
    return original_data

inverse_transformed_data = inverse_transform(q_matrix, transformed_data, Mean)
print("Inverse Transformed Data :\n", inverse_transformed_data)
```

```python
inverse_transformed_data.shape
```

```python
def reconstruction_error(original_data, reconstructed_data):
    error = np.linalg.norm(original_data - reconstructed_data) / np.linalg.norm(original_data)
    return error

# Calculate reconstruction error for the current transformation
error = reconstruction_error(data, inverse_transformed_data)
print("reconstruction Error:", error)
```

```python
def euclidean_distance_matrix(data):
    n = data.shape[0]
    distance_matrix = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            distance_matrix[i, j] = np.linalg.norm(data[i] - data[j])
    return distance_matrix
```

```python
from sklearn.preprocessing import StandardScaler
data_scaled = StandardScaler().fit_transform(data)
distance_matrix = euclidean_distance_matrix(data_scaled)
print("Euclidean Distance Matrix:\n", distance_matrix)
```

```python
def eps_neighborhood_all(data, epsilon):
    distance_matrix = euclidean_distance_matrix(data)
    print(distance_matrix)
    neighborhoods = []
    
    for i in range(distance_matrix.shape[0]):
        neighborhood_indices = np.where(distance_matrix[i] <= epsilon)[0]
        neighborhoods.append(neighborhood_indices)
    
    return neighborhoods

epsilon = 2  

neighborhoods = eps_neighborhood_all(data_scaled, epsilon)
print(len(neighborhoods))
for i, neighborhood in enumerate(neighborhoods):
    print(f"Eps-neighborhood of point {i}:", neighborhood)
```

```python
def classify_points(neighborhoods, min_pts):
    n = len(neighborhoods)
    core_points = []
    border_points = []
    noise_points = []
    
    # Determine core points
    for i in range(n):
        if len(neighborhoods[i]) >= min_pts:
            core_points.append(i)
    
    # Determine border and noise points
    for i in range(n):
        if i not in core_points:
            # Check if the neighborhood contains at least one core point
            if any(point in core_points for point in neighborhoods[i]):
                border_points.append(i)
            else:
                noise_points.append(i)
    
    return core_points, border_points, noise_points

min_pts = 5  
core_points, border_points, noise_points = classify_points(neighborhoods, min_pts)

# Output results
print("\nCore Points:", core_points)
print("Border Points:", border_points)
print("Noise Points:", noise_points)
```

```python
def dbscan_clustering(data, epsilon, min_pts):
    
    visited = set()  # Tracks visited points
    clusters = {}  # Dictionary to store clusters
    cluster_id = 0
    
    for point_idx in range(len(data)):
        if point_idx in visited or point_idx in noise_points:
            continue  # Skip already visited or noise points
        
        if point_idx in core_points:
            # Start a new cluster
            cluster_id += 1
            clusters[cluster_id] = []
            queue = [point_idx]  # Initialize a queue with the core point
            
            while queue:
                current_point = queue.pop(0)
                
                if current_point not in visited:
                    visited.add(current_point)
                    clusters[cluster_id].append(current_point)
                    
                    # Add neighbors to the cluster if they are core or border points
                    for neighbor in neighborhoods[current_point]:
                        if neighbor not in visited and neighbor not in queue:
                            if neighbor in core_points or neighbor in border_points:
                                queue.append(neighbor)
    
    return clusters, noise_points
clusters, noise_points = dbscan_clustering(data_scaled, epsilon, min_pts)

# Output results
print("\nClusters:")
for cluster_id, points in clusters.items():
    print(f"Cluster {cluster_id}: {points}")

print("\nNoise Points:", noise_points)
```
